```python
normals=pd.Series(np.random.normal(size=20))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
var=pd.DataFrame({'normal':np.random.normal(size=10),
                 'gamma':np.random.gamma(1,size=10),
                 'poisson':np.random.poisson(size=10)})
print(var)
var.cumsum(0).plot()
```

```python
var.cumsum(0).plot(subplots=True)
```

```python
var.cumsum(0).plot(secondary_y='normal')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(4,12))
for i,v in enumerate(['normal','gamma','poisson']):
    var[v].cumsum(0).plot(ax=axes[i],title=v)
axes[0].set_ylabel('cumulative sum')
```

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Greens')
plt.colorbar();
```

```python
x = np.linspace(0, 10, 100)
print(x)
dy = 0.1
y = np.sin(x) + dy * np.random.normal(100)

plt.errorbar(x, y, yerr=dy, fmt='*',ecolor='red');
```
